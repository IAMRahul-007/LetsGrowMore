### **IMPORTING LIBRARIES**

```python
import cv2
from PIL import Image
import matplotlib.pyplot as plt
%matplotlib inline
```

### **LOADING THE IMAGE**

```python
img = cv2.imread("im.jpg")
img = cv2.resize(img,(800,800))
i_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(i_rgb)
```

### **CONVERTING THE IMAGE TO GREYSCALE**

```python
from IPython.display import display
```

```python
grey_img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
display(Image.fromarray(grey_img))
```

### **CONVERTING THE IMAGE TO NEGATIVE**

```python
invt_img = cv2.bitwise_not(grey_img)
display(Image.fromarray(invt_img))
```

### **SMOOTHENING THE IMAGE**

```python
blur_img = cv2.GaussianBlur(invt_img, (5,5), 0, 0)
display(Image.fromarray(blur_img))
```

### **CREATING THE PENCIL SKETCH FROM IMAGE**

```python
invt_blur_img = 255 - blur_img
sketch = cv2.divide(grey_img, invt_blur_img, scale= 250.0)
display(Image.fromarray(sketch))
```

### **CONCLUSION - COMPARING THE SKETCH AND ORIGINAL IMAGE**

```python
i_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(i_rgb)
display(Image.fromarray(sketch))
```
